### Vector Space

A vector space is like a playground for vectors where we could perform mathematical operations freely.

Vector Space (Linear Space) Rule: 

A set of objects (vectors, matrices, tensors) forms a space if:

- Addition is defined → sum of any two objects stays in the set. 
- Scalar multiplication is defined → multiplying an object by a scalar stays in the set. 
- Supporting Axioms (must also hold for a true vector space)
    - Associativity: (u + v) + w = u + (v + w)
    - Commutativity: u + v = v + u
    - Zero vector exists: v + 0 = v
    - Additive inverse: v + (-v) = 0
    - Distributivity: k(u + v) = ku + kv and (k + m)v = kv + mv
    - Multiplicative identity: 1*v = v

Hence,    
1D array (vector) → forms a 1D vector space.    
2D array (matrix) → forms a 2D matrix space.    
nD array (tensor) → forms an nD tensor space.

- Purpose: Words that appear often in a document but rarely across the corpus get higher weight.

# A. Regression Algorithms

## 1. Linear Regression

`As an AI Automation Consultant, Linear Regression is a supervised learning method used to predict continuous values from input features using a linear relationship (straight line)`

- Predicts numeric output (y) from input features (X).
- Uses a linear equation: y = w₁x₁ + w₂x₂ + … + wₙ*xₙ + b.
- Weights (w) and bias (b) are learned from training data.
- Goal is to minimize difference between predicted value (ŷ) and actual values (y).
 
 *Adjustable Parameters :*   
*Weights (w) → influence of each feature on prediction.*.  
*Bias (b) → starting point or baseline adjustment to better fit the data.*

**Types**

1. Simple Linear Regression.
    - One feature (X)
    - Predicts numeric output (ŷ)

2. Multiple Linear Regression. 
    - Two or more features (X₁, X₂, … Xₙ)
    - Predicts numeric output (ŷ)

### Which line to choose ? 

In Linear Regression, the goal is to plot the “best fit line”: The line that minimizes the distance (error) between actual data points and the predicted values.

**Methods :**

1. Least Sqaured Loss Function (LSLF) : 
    - LSLF is the sum of squared differences between actual values and predicted values.

So instead: measure distance of points from line, square it, sum it, and pick the line that scores lowest.




## 2. Logistic (Logit) Regression

### Logistics Regression

`As an AI Automation Consultant, Logistic Regression exists because some problems require predicting categories (yes/no, 0/1, class labels) instead of continuous numbers.`

- It is a supervised learning algorithm for classification. 
- Predicts probability of a class between 0 and 1 using the logistic (sigmoid) function (σ(z)) for Binary Class and Softmax for Multi Class.
- Model still has weights (w) and bias (b) learned via Gradient Descent.
- Loss function: Binary Cross-Entropy (Log Loss). 
- Decision rule:
    - If σ(z) ≥ 0.5 → class 1.
    - If σ(z) < 0.5 → class 0.

### Confusion Matrix

### Evaluation Metrics

`As an AI Automation Consultant, evaluation metrics exist because minimizing loss during training does not guarantee that a model will perform well on unseen, real-world data.`

- Used to judge model performance after training.
- Loss helps learning; metrics help evaluation.

*Evaluation metrics are mainly used during the validation phase to assess and improve model performance, and later during inference to monitor how the deployed model behaves on real-world data.*

### Label Based Evalution Metrics 

These metrics are used in supervised learning, and they differ based on the task type: classification uses decision-based metrics, while regression uses error-based metrics.

#### 1. Decision Based Metrics (Classification Supervised Learning)

- Measure correctness of decisions, not numerical error.
- Built using:
    - True Positive (TP)
    - True Negative (TN)
    - False Positive (FP)
    - False Negative (FN)

**Types**

1. Accuracy

- Accuracy measures out of all decisions made by a classification model, how many were correct?
- Can be misleading for imbalanced datasets.
- It is calculated as : 
    - Accuracy = (TP + TN) ÷ (TP + TN + FP + FN)
- Use when: Classes are balanced and all mistakes are equally costly.

2. Precision 

- Precision measures when the model predicts “Yes/Positive,” how often is it actually correct?
- It is calculated as : 
    - Precision = TP ÷ (TP + FP)
- Use when: False positives are costly.

3. Recall

- Recall measures out of all actual positives, how many did the model correctly identify?
- It is calculated as : Recall = TP ÷ (TP + FN).
- Use when: False negatives are costly.

4. F1 Score

- F1 Score exists to answer: how well the model balances being correct (Precision) and catching positives (Recall).
- It is especially important when both false positives and false negatives matter.
- F1 = 2 × (Precision × Recall) ÷ (Precision + Recall)
- High F1 → model has good balance between Precision and Recall.
- Use when: You want a balance between precision and recall, especially for imbalanced datasets.

5. ROC Curve 

- ROC : Receiver Operating Characteristic Curve.
- The ROC Curve exists to show how well a classification model separates classes (computer distinguishes Yes vs No) across all possible thresholds, not just the default 0.5.
- Plots:
    - True Positive Rate (Recall) on Y-axis (Sensitivity) : 
        - TPR = TP ÷ (TP + FN)
        - how many actual positives were correctly identified.
    - False Positive Rate (FPR) on X-axis (Specifity) : 
        - FPR = FP ÷ (FP + TN) 
        - how many actual negatives were incorrectly classified as positive. 
- Use when: You want to compare models or choose the optimal threshold and classes are imbalanced.

6. AUC Curve

- AUC : Area Under Curve.
- AUC exists to summarize the ROC Curve into a single number that indicates how well a classification model separates positives from negatives across all thresholds.
- Higher AUC → better model at distinguishing classes.
- 0.5 → random, 1 → perfect.
- Use when: You want to compare models or choose the optimal threshold and classes are imbalanced.

7. Misclassification Error

- Measures out of all the decisions, how many incorrect outputs are predicted by the machine.
- It is calculated as : Misclassification Error = (FP+FN) ÷ (TP+TN+FP+FN).

# D. Gradient (Slope) Descent Algorithm 

Gradient Descent is an optimisation algorithm used in supervised learning model to find optimal parameters to make accurate predictions. 

`As an AI Automation Consultant, Gradient Descent exists because once we define a loss (like MSE), we need a systematic, efficient method to find the values of weights and bias that minimize that loss.`

- It is an optimization algorithm.
- Purpose: minimize any loss function.
- It works by:
    - Computing the gradient (slope) of the loss w.r.t. parameters (w,b).
    - Updating weights and bias in the opposite direction of the gradient.
        - If the slope (gradient) is positive, the parameter value is decreased
        - If the slope (gradient) is negative, the parameter value is increased.   
        *This is because Gradient Descent always moves downhill to minimize the loss.*      
- Happens iteratively: Predict initial values → compute loss → compute gradient → update parameters.

*It is what actually drives learning in most ML models.*

*Local minima are points where gradient = 0 but loss is not lowest; our model may get stuck there. Use stochastic methods, momentum, or multiple initializations to escape and reach the global minimum.*

### Types of Gradient Descent

`As an AI Automation Consultant, different types of Gradient Descent exist because updating model parameters efficiently depends on the size of the dataset and how often we calculate gradients.`

- Batch Gradient Descent
    - Uses all training data per iteration to compute the gradient
    - Pros: stable and accurate step toward minimum
    - Cons: slow for large datasets
- Stochastic Gradient Descent (SGD)
    - Uses one training sample per iteration
    - Pros: faster, can escape local minima
    - Cons: noisy updates → less stable convergence
- Mini-Batch Gradient Descent
    - Uses a small batch of samples per iteration
    - Pros: balance between speed and stability
    - Most commonly used in practice

*It’s like deciding how much information to consider before adjusting your move downhill.*

### Learning Rate (Step Size) (α)

- Hyperparameter controlling step size in Gradient Descent.
- Determines how much weights (w) and bias (b) are updated per iteration.
- Small α → slow learning.
- Large α → unstable or diverging learning; however, initial steps could be large to expedite the process.

*Proper, Efficient tuning is key to fast and stable training.*

Gradient Descent updates weights (w) and bias (b) based on the computed gradient of the loss:
- The sign of the gradient determines whether to add or subtract from the parameter.
- The magnitude of the update is controlled by the step size (learning rate).
- The type of Gradient Descent (Batch, SGD, Mini-Batch) decides how the gradient is computed.

# Image Processing

`As an AI Automation Business Consultant, we handle different types of data; hence, when data is visual (documents, images, scans etc), Image Processings helps to prepare such data for ML models so that they could learn better and produce outputs with accuracy.`

`Image Processing for visual data is similar to Digital Signal Processing for signals in which data is being processed and prepared for our ML Models.`

- Image Processing treats an image as a 2D signal (matrix) of pixel values.
- Each cell would represent a pixel and each pixel would have magnitude (intensity or color value).
- Image Processing is usually performed before machine learning so models receive clean and structured input data.
- Goal: enhance quality or extract useful patterns.

### Techniques

1. Convolution

    Convolution is the core operation that lets AI detect patterns in images and signals.
    - It is a mathematical operation between an input (signal/image) and a kernel (filter).
    - The kernel (filter) would slide across the input to find patterns and thus producing feature map (output).
    - Every kernel is a filter and each filter detects a small pattern.
    - Convolution converts raw data into feature maps that are easier for AI models to learn from.
    - Parameters that matter: kernel size, stride, padding.
        - Kernel matrix must be square and of odd dimensions, minimum meaninful Kernel Size is 3X3.
        - Padding is adding extra pixels across the border of the image before convolution to avoid data loss at edges. 
            - Replication Padding
            - Zero Padding
    - Operator = * (*this is an operator for convolution between matrices and not multiplication.*) 
 
2. Spatial Filtering

    Spatial filtering improves images by modifying each pixel (sharpening, smoothening...) using information from its neighbors.
    Smoothening Filter : 
    - Used for blur and noise reduction.



3. Geometric Transformation

    Geometric transformations change where pixels are located, not their values. They are essential for alignment and normalization, so AI systems receive consistent, well-aligned inputs.

    Transformation Types :    
        - Linear Transformation = Reshape Images (Shearing, Rotation, Scaling, Reflection).    
        - Affine Transformation = Reshape and Reposition Images (Linear Transformation + Translation). 

    Techniques :    

    A. Shearing

    - Transformation that slants (tilts) an image.
    - Horizontal Shear : Across x-axis.
    - Vertical Shear : Across y-axis. 

    B. Rotation

    - Tranformation that rotates an image across a fixed point. 
    - Clockwise Rotations.
    - Anti-Clockwise Rotation.

    C. Scaling 

    - Transformation by multiplying pixel coordinates by scaling factors.
    - Enlarged/ Stretched.
    - Compressed/ Shrinked.

    D. Translation 

    - Transformation to shift an image in x and/or y direction.

    E. Reflection 

    - Transformation that flips an image across an axis. 
    - Horizontal flip across y-axis.
    - Vertical flip across x-axis. 
    - Flip across Origin.    

4. Intensity Transformation

    Unlike Geometric Transformation, Image processing technique in which pixel values would be changed, and pixel positions would remain unchanged. It helps reveal important details in low-quality images.

    Types : 
    1. Photographic Negative.
    - A gray-level transformation that reverses the intensity of every pixel in an image.
    - Dark becomes bright; bright becomes dark.

    2. Gamma Transformation.
    - Transformation used to adjust brightness and contrast in an image.
    - Gamma (variable), if > 1 compresses the pixels, making it look darker; however, if < 1 enlarge the pixels, making it look brighter.

    3. Logarithmic Transformation.
    - Transformation used to balance dark pixels by enlarging them and light pixels by compressing them. 

    4. Contrast Stretching Transformation. 
    - Transformation used to make dark pixels appear darker and light pixels appear lighter.
