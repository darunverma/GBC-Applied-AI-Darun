# MACHINE LEARNING FUNDAMENTALS


### Machine Learning (ML)

It is a part of AI where systems learn patterns from data to make predictions or decisions without being explicitly programmed for every rule.

`As an AI Automation Consultant, ML is not about models — it’s about decisions at scale.`

### Tradtional Programming VS Machine Learning

| Traditional Programming | Machine Learning      |
| ----------------------- | --------------------- |
| Rules + Data → Output   | Data + Output → Rules |
| Rigid                   | Adaptive              |
| Breaks easily           | Improves with data    |

Traditional Programming : Learns from the "Rules" provided to produce Output.  
Machine Learning : Learns from the Data Samples and Output given to generate "Rules".

### Reason for Existence of Machine Learning 

`As an AI Automation Consultant, Machine Learning exists because many real-world business problems cannot be solved with fixed rules, yet they still require consistent, scalable decisions.
Humans handle these well — but humans don’t scale.`

ML solves problems where:
* Explicit rules are unknown, incomplete, or too complex.
* Patterns exist inside data rather than logic.
* The environment changes over time.
* Performance improves with experience.

So instead: You let the computer learn from past examples, just like a human learns from experience.

### Building ML Intuition

`As an AI Automation Consultant, building ML intuition is about understanding how data leads to decisions, rather than memorizing algorithms.`

* Focus on how data represents the world
* Look for patterns and how they generalize
* Observe how models learn from examples and improve over time
* Pay attention to errors and what they indicate

So instead: You look at examples, notice patterns, and see how the system makes predictions, rather than memorizing formulas.

*If I understand how ML learns from data and improves over time, I can use that understanding to build my ML intuition.*

### Phases in Machine Learning

`As an AI Automation Consultant, Machine Learning works in distinct phases: one where the system learns from data, and another where it applies that learning to make predictions.`

- Training Phase:
    * Model learns patterns from data

# MACHINE LEARNING CONCEPTS

### Lables (y)

`As an AI Automation Consultant, labels are the correct outputs provided with training data in supervised learning.`

- Represent the target variable (y).
- Used to calculate errors during training.
- Help the model learn from examples.

### Features (X)

`As an AI Automation Consultant, features are the input variables provided to the model to learn patterns or make predictions.`

- Represent the data points used for learning (X).
- Can be numeric, categorical, or text-based.
- Help the model predict labels or identify patterns.
- Quality of features directly affects model performance.

### ML Model

`As an AI Automation Consultant, a Machine Learning Model is a system that has learned patterns from data and can make predictions on new inputs.`

- Created after the training phase.
- Adjusts internal parameters to minimize error.




### Linear Regression

`As an AI Automation Consultant, Linear Regression is a supervised learning method used to predict continuous values from input features using a linear relationship (straight line)`

- Predicts numeric output (y) from input features (X).
- Uses a linear equation: y = w₁x₁ + w₂x₂ + … + wₙ*xₙ + b.
- Weights (w) and bias (b) are learned from training data.
- Goal is to minimize difference between predicted value (ŷ) and actual values (y).
 
 *Adjustable Parameters :*   
*Weights (w) → influence of each feature on prediction.*.  
*Bias (b) → starting point or baseline adjustment to better fit the data.*

**Types**

1. Simple Linear Regression.
    - One feature (X)
    - Predicts numeric output (ŷ)

2. Multiple Linear Regression. 
    - Two or more features (X₁, X₂, … Xₙ)
    - Predicts numeric output (ŷ)


### Model Training (Linear Regression)

- Model Training is the process where a model learns optimal values of the adjustable parameters weights (w) and bias (b) with the help of training data (X) and (y).
- Optimal values of weights (w) and bias (b) are the values that minimize the loss (error) between the actual output (y) and the predicted output (ŷ), where the loss is computed using a loss function (often based on the difference between y and ŷ).

### Empirical Risk Minimisation (ERM) (Goal)

- Learning strategy used during model training.
- Focuses on minimizing error on training data.

### Loss

- Signify how wrong the model is performing on training data.
- Computed per data point (feature).
- Calculated using:
    - Actual value (y)
    - Predicted value (ŷ)

### Mean Squared Error (MSE)

- It is a loss function used in regression.
- It measures the average of squared differences between actual and predicted values.
- Reason for Squaring (y − ŷ)²:
    - Removes negative signs.
    - Penalizes large errors more heavily.

*So, MSE tells how bad predictions are on average of model, with extra punishment for big errors.*

MSE is computed iteratively during training, and the model updates weights (w) and bias (b) in each iteration to minimize the MSE loss.

### Gradient (Slope) Descent

`As an AI Automation Consultant, Gradient Descent exists because once we define a loss (like MSE), we need a systematic, efficient method to find the values of weights and bias that minimize that loss.`

- It is an optimization algorithm.
- Purpose: minimize the loss function.
- It works by:
    - Computing the gradient (slope) of the loss w.r.t. parameters (w,b).
    - Updating weights and bias in the opposite direction of the gradient.
        - If the slope (gradient) is positive, the parameter value is decreased
        - If the slope (gradient) is negative, the parameter value is increased.   
        *This is because Gradient Descent always moves downhill to minimize the loss.*      
- Happens iteratively: Predict initial values → compute loss → compute gradient → update parameters.

*It is what actually drives learning in most ML models.*

### Types of Gradient Descent

`As an AI Automation Consultant, different types of Gradient Descent exist because updating model parameters efficiently depends on the size of the dataset and how often we calculate gradients.`

- Batch Gradient Descent
    - Uses all training data per iteration to compute the gradient
    - Pros: stable and accurate step toward minimum
    - Cons: slow for large datasets
- Stochastic Gradient Descent (SGD)
    - Uses one training sample per iteration
    - Pros: faster, can escape local minima
    - Cons: noisy updates → less stable convergence
- Mini-Batch Gradient Descent
    - Uses a small batch of samples per iteration
    - Pros: balance between speed and stability
    - Most commonly used in practice

*It’s like deciding how much information to consider before adjusting your move downhill.*


### Hyperparameters

Hyperparameters are parameters set before training to control the learning behaviour. These are not updated by the training process unlike weights (w) and bias (b).


### Learning Rate (Step Size) (α)

- Hyperparameter controlling step size in Gradient Descent.
- Determines how much weights (w) and bias (b) are updated per iteration.
- Small α → slow learning.
- Large α → unstable or diverging learning; however, initial steps could be large to expedite the process.

*Proper, Efficient tuning is key to fast and stable training.*


Gradient Descent updates weights (w) and bias (b) based on the computed gradient of the loss:
- The sign of the gradient determines whether to add or subtract from the parameter.
- The magnitude of the update is controlled by the step size (learning rate).
- The type of Gradient Descent (Batch, SGD, Mini-Batch) decides how the gradient is computed.



### Ockham’s Razor & Generalization

- Principle: simpler models generalize better.
- Avoid unnecessary complexity in model design.
- Helps prevent overfitting
- Guideline for choosing models that perform well on new, unseen data.

*In AI consulting, applying Ockham’s Razor ensures models are robust and scalable for business applications, instead of being overly tuned to past data.*

### KISS Principle

Keep It Simple, Stupid.   

Simplicity leads to reliability, easier maintenance, and better generalization.

### Goldilocks Principle

The Goldilocks Principle refers to finding the “just right” balance in ML settings:
- Model complexity: too simple → underfitting, too complex → overfitting
- Learning rate: too small → slow learning, too large → unstable learning
- Number of features: too few → underfit, too many → overfit


### Dataset Splitting

Dataset splitting divides available data into:
- Training set → used to train the model.
- Validation set → used to tune hyperparameters and prevent overfitting.
- Test set → used to evaluate final model performance.
- Typical ratios:
    - Training: 70–80%
    - Validation: 10–15%
    - Test: 10–15%

Ensures model is not evaluated on data it has already seen

### Regularisation : 

`As an AI Automation Consultant, Regularization exists because models can overfit—learn training data too perfectly, including noise—so we need a way to keep the model simple and generalizable.`

- Technique to reduce overfitting.
- Adds penalty to loss function based on weight size (influence) (w).
- Types of Explicit Regularisation:
    - L1 (Lasso): sum of absolute weights → can remove unimportant features.
    - L2 (Ridge): sum of squared weights → shrinks weights.
    - Elastic Net : Benefits of both Lasso (feature selection) and Ridge (shrinkage).
- Total Loss = Original Loss + Regularization Term.
- Encourages simpler, generalizable models.

*Lambda (λ) exists because we need a control knob to decide how much we penalize large weights when applying regularization.*
- Effect:
    - Large λ → stronger penalty → smaller/sparser weights → simpler model
    -   Small λ → weaker penalty → model behaves closer to original (less regularized)






### Logistics Regression

`As an AI Automation Consultant, Logistic Regression exists because some problems require predicting categories (yes/no, 0/1, class labels) instead of continuous numbers.`

- It is a supervised learning algorithm for classification. 
- Predicts probability of a class between 0 and 1 using the logistic (sigmoid) function (σ(z)).
- Model still has weights (w) and bias (b) learned via Gradient Descent.
- Loss function: Binary Cross-Entropy (Log Loss). 
- Decision rule:
    - If σ(z) ≥ 0.5 → class 1.
    - If σ(z) < 0.5 → class 0.

One of them is: 

1. Early Stopping :

### Evaluation Metrics

`As an AI Automation Consultant, evaluation metrics exist because minimizing loss during training does not guarantee that a model will perform well on unseen, real-world data.`

- Used to judge model performance after training.
- Loss helps learning; metrics help evaluation.

*Evaluation metrics are mainly used during the validation phase to assess and improve model performance, and later during inference to monitor how the deployed model behaves on real-world data.*

### Label Based Evalution Metrics 

These metrics are used in supervised learning, and they differ based on the task type: classification uses decision-based metrics, while regression uses error-based metrics.

#### 1. Decision Based Metrics (Classification Supervised Learning)

- Measure correctness of decisions, not numerical error.
- Built using:
    - True Positive (TP)
    - True Negative (TN)
    - False Positive (FP)
    - False Negative (FN)

**Types**

1. Accuracy

- Accuracy measures out of all decisions made by a classification model, how many were correct?
- Can be misleading for imbalanced datasets.
- It is calculated as : 
    - Accuracy = (TP + TN) ÷ (TP + TN + FP + FN)
- Use when: Classes are balanced and all mistakes are equally costly.

2. Precision 

- Precision measures when the model predicts “Yes/Positive,” how often is it actually correct?
- It is calculated as : 
    - Precision = TP ÷ (TP + FP)
- Use when: False positives are costly.

3. Recall

- Recall measures out of all actual positives, how many did the model correctly identify?
- It is calculated as : Recall = TP ÷ (TP + FN).
- Use when: False negatives are costly.

4. F1 Score

- F1 Score exists to answer: how well the model balances being correct (Precision) and catching positives (Recall).
- It is especially important when both false positives and false negatives matter.
- F1 = 2 × (Precision × Recall) ÷ (Precision + Recall)
- High F1 → model has good balance between Precision and Recall.
- Use when: You want a balance between precision and recall, especially for imbalanced datasets.

5. ROC Curve 

- ROC : Receiver Operating Characteristic Curve.
- The ROC Curve exists to show how well a classification model separates classes (computer distinguishes Yes vs No) across all possible thresholds, not just the default 0.5.
- Plots:
    - True Positive Rate (Recall) on Y-axis (Sensitivity) : 
        - TPR = TP ÷ (TP + FN)
        - how many actual positives were correctly identified.
    - False Positive Rate (FPR) on X-axis (Specifity) : 
        - FPR = FP ÷ (FP + TN) 
        - how many actual negatives were incorrectly classified as positive. 
- Use when: You want to compare models or choose the optimal threshold and classes are imbalanced.

6. AUC Curve

- AUC : Area Under Curve.
- AUC exists to summarize the ROC Curve into a single number that indicates how well a classification model separates positives from negatives across all thresholds.
- Higher AUC → better model at distinguishing classes.
- 0.5 → random, 1 → perfect.
- Use when: You want to compare models or choose the optimal threshold and classes are imbalanced.
 

***Classification Threshold***
- It is a probability cutoff (default often 0.5).
- Model Predictions ≥ threshold → Positive
- Model Predictions < threshold → Negative
- Changing the threshold affects:
    - Precision (higher threshold → fewer false positives).
    - Recall (lower threshold → fewer false negatives).

# Feature (X) Selection


`A labeled dataset consists of data samples represented as rows, features represented as columns whose values form column vectors, and a separate label column vector where each data sample has a corresponding output value.`

- Feature Selection is the process of choosing a subset of relevant features from the original feature set.
- Purpose:
    - Reduce overfitting & dimensionality
    - Improve model performance
    - Reduce training time
    - Improve interpretability
    - Reduce complexiety
- It removes irrelevant, redundant, or noisy features
- It is different from Feature Engineering:
    - Feature Engineering → create or transform features
    - Feature Selection → choose the best existing features


**Types (***for Supervised Learning***)**

1. Filter
    - Select features before model training.
    - It evaluates each feature (X) independently.
    - Model-independent.
2. Wrapper
    - Select subset of features > train the model using algorithm > checks model performance > and keeps the best-performing set.
    - Evaluate performance for different feature subsets
    - Model-dependent and computationally expensive
3. Embedded
    - Feature selection happens during model training
    - Uses model’s internal learning process

##### A. Filter Method

***Clean the data first, then teach the computer.***

*Filter methods use ranking techniques where features are assigned statistical scores based on their correlation or dependency with the outcome variable, and features are selected using rank ordering.*

1. Pearson Corelation (Feature-Label relevance)

-   Measures linear relationship between a feature (X) and label (y).
-   Range: –1 to +1
    - +1 → perfect positive linear relationship.
    - –1 → perfect negative linear relationship.
    - 0 → no linear relationship.
-   Absolute values of Pearson Corelation is being used to determine the relationship for filteration.

2. Chi Squared (Feature-Label relevance)

- Measures statistical dependence between:
    - Categorical feature (X).
    - Categorical label (y).
- Range: 0 to Infinity
    - High Chi Squared score → strong relationship
    - Low Chi Squared score → weak or no relationship

3. Mutual Information Gain (Feature-Leabel relevance)

- Measures the shared information between a feature (X) and the label (y).
- Works with:
    - Numeric features
    - Categorical features
- Range: 0 to Infinity
    - High Mutual Information Gain → stronger dependency
    - Low Mutual Information Gain→ weak dependency

4. Fishers Score (Feature-Feature relevance)

- Measures how well a feature separates classes.
- Range: 0 to Infinity
    - 0 → No class separability (useless feature)
    - Low value → Weak separation between classes
    - High value → Strong separation between classes

5. Corelation Coeeficient with Heatmap (Redundancy detection)

- Measures linear relationship between pairs of numerical features (feature to feature).
- Range: –1 to +1
    - +1 → perfect positive linear corelation.
    - –1 → perfect negative linear corelation.
    - 0 → no linear corelation.

    So instead: If two inputs move together strongly, keep one and drop the other.

6. Variance Threshold (Feature variability)

- Removes features whose variance (Spread of values) is below a set threshold.
- Measure how much a feature varies across samples (Does this feature change at all ?)
- Only works on numerical features.
- Independent of label (unsupervised filtering).
- Threshold is set depending on the dataset: 
    - 0 → remove constant features
    - >0 → remove features with low variability

7. Mean Absolute Difference (MAD) (Feature variability)

- Measure Avg distance from mean (On average, how far do values deviate ?) 
- Range: 0 → ∞
    - 0 → all values identical
    - Higher → more spread across samples
- Often used as an unsupervised filter.
- Works on numerical features

8. Dispersion Ratio (Feature variability)

- Quantify how spread out a feature is relative to its mean (Relative Spread) (Is the variability large compared to the feature’s scale (mean)?)
- Range: 0 → ∞
    - 0 → all values identical
    - Higher → more spread across samples
- Works for numerical features.


##### B. Wrapper Method

***You let the computer learn from past examples, try different inputs, and keep the set that predicts best.***

1. Forward Feature Selection

    Steps : 
- Start with an empty set of features.
- Add the feature that gives the best performance improvement.
- Repeat until adding more features doesn’t improve performance.

2. Backward Feature Elimination

    Steps :
- Start with all features.
- Remove the feature whose removal improves or least affects performance.
- Repeat until removing more features hurts performance.

3. Exhaustic Feature Selection 

    Steps : 
- Generate all possible feature subsets.
- Train a model on each subset.
- Evaluate performance.
- Select the subset with the best performance.

4. Recursive Feature Elimination 

    Recursively removes the least important feature based on the model’s ranking until the desired number of features is reached.    

    Steps : 
- Train a model on all features.
- Rank features based on importance.
- Remove the least important feature.
- Retrain model and repeat.
- Continue until desired feature set is obtained.

*BFE = “remove least useful feature, check performance, repeat”*   
*RFE = “rank features using model importance, remove the least important, retrain, repeat until the best subset remains”*
*Even though both are iterative, RFE is more systematic and model-aware, while BFE is simpler and less precise.*

##### C. Embedded Method

***You let the computer learn and pick the important features at the same time, without doing extra feature selection.***

1. Lasso Regression (L1 Regularization)

- It adds penalty = sum of absolute coefficients to the loss function.
- Coefficients of unimportant features are shrunk to 0 → feature removed.

2. Random Forest - Tree Based

    Random Forest trains many independent trees, sees which features help most (in reducing loss), and automatically highlights important features.

- Rank features → remove low-importance ones

3. Light GBM

    LightGBM trains sequential trees, finds which features reduce error most, and highlights important features automatically.

- It is a gradient boosting framework that builds trees and calculates feature importance during model training.
- Handles categorical features efficiently.
- Rank features → remove low-importance ones.

# Data (X) Preprocessing

`As an AI Automation Business Consultant, Good preprocessing ensures your AI automation models learn correctly from clean, structured, and relevant data.`

- Data Preprocessing is the set of techniques applied to raw data to make it suitable for machine learning models.
- Business relevance :
    - Ensures reliable automation decisions.
    - Improves model accuracy and stability.



### Methods (*as taught in ML class*) :

##### A. Value Simplification / Bucketing *(Turn raw numbers into meaningful ranges or decisions.)*


1. Binarisation

    Binarisation is a data preprocessing technique which is used to convert numerical or categorical values into binary vector (0,1/ True, False/ Yes,No).
    - Often done using a threshold.
    - Special case of discretisation with only two bins.
    - Aligns data with decision-based workflows.

    *Tools :*
    1. Binarizer()

2. Discretisation 

    - Discretisation is a data preprocessing technique which is used to convert continuous numerical values into discrete ranges/ categories called bins.
    - Reduces noise and simplifies decision-making.
    - Useful for both supervised and unsupervised learning.  

    *Tools :*
    - KBinsDiscretizer()



##### B. Categorical Data Representation *(Make categorical data understandable to ML models.)*

1. On-Hot Encoding

    One-Hot Encoding is a data preprocessing technique that converts n categorical features into n binary vectors (0 or 1) so that each category has its own column.
    - Disadvantage = One-Hot Encoding can create too many columns, especially when a categorical feature has high number of categories. 

    *Tools :*
    1. OneHotEncode()
    2. getdummies(), *available in pandas library that works the same; however, names of newly created columns are originalcolumnname_category.*


2. Label Encoding
    
    Label Encoding is a data preprocessing technique that converts a categorical feature into numerical vector (0,1,2,3,4......), assigning values based on ascending alphabetical order.

    *Tools :*
    1. LabelEncoder()


        *So instead: use label encoding only when order exists or the model can handle categorical splits safely. Trees are safe with Label Encoding; linear and distance-based models are not.*


##### C. Value Transformation / Scale Adjustment *(Ensure numbers are comparable and models behave correctly.)*

1. Feature Scaling 

    - Feature Scaling ensures numerical features are comparable by adjusting them on a common scale so models learn properly.
    - Methods:
        - Min-Max Scaling → scale to [0,1]
        - Standardization → mean=0, std=1, mean centered.
        - Robust Scaling → median-centered, resistant to outlier.

    *Feature rescaling is applied when some features have a much higher range of values than others, so that all features contribute fairly to the model.*

    *Tools :*
    1. MinMaxScaler()
    2. scale()
    3. Standardscaler()
    4. Pipeline()  
    ...

2. Normalisation

    - Normalization ensures inputs are comparable in direction.
    - Scales each row independently.
    - It transforms numeric sample to a common scale, usually [0,1] or [-1,1].
    - Magnitude = norm = 1.

    So instead: choose normalization when you care about similarity or patterns between samples, not absolute values.

    *Tools :*
    1. Normalize()



##### D. Handling Missing Data *(Make incomplete data usable without deleting it.)*

Imputation

- Imputation is a data preprocessing technique of filling in missing values in a dataset.
- Missing data can break models or bias results; hence, instead of deleting data, we estimate reasonable values so learning can continue.

*Types :*

1. Univariate Imputation (SimpleImputer) :

- Imputes one feature at a time, independently.
- Uses simple statistics: mean, median, mode, constant.
- Does not consider relationships between features.

    *Tools :*
            - SimpleImputer()

2. Multivariate Imputation (IterativeImputer) :

- Imputes missing values using relationships between features.
- Treats each feature as a target and predicts it from others.
- Iterative process until values stabilize.
- More accurate but computationally heavier.
- Regressor model (predicts a continuous numeric value using historical data) is used to predict the missing values.

    *Tools :*
        - IterativeImputer()

# Machine Learning Algorithms

Machine Learning, a subset of Artificial Intelligence, enables systems to learn from data to predict outcomes on unseen information. While traditional programming relies on manually coded logic, Machine Learning allows systems to derive their own rules from data. These internal rules are governed by specialized algorithms developed by researchers and engineers which we call Machine Learning Algorithms or ML Algorithms.

## 1. Decision Trees

`In AI automation, many business decisions already happen as rules: if this happens, then do that. Decision Trees learn these rules automatically from data, making them ideal for automating approvals, risk checks, lead qualification, customer segmentation, and operational decisions.`

##### Core Idea

- A Decision Tree is a supervised learning algorithm.
- It works by splitting data into smaller and purer (homogeneous splits) groups step by step.
- Each split is based on a feature that best separates the data.
- The process continues until a final decision (output) is reached. 

##### How a Decision Tree Learns

1. Start with the Entire Dataset
    - The algorithm begins with all training data at the root node.
    - At this stage, data contains mixed outcomes and is not well organized.

2. Evaluate All Features
    - Each feature is tested as a possible splitting point.
    - The algorithm checks how well a feature can divide the data into cleaner groups.

3. Choose the Best Split
    - The best split is the one that reduces impurity the most.
    - Impurity measures how mixed the data is.
    - Common impurity measures:
        - Gini Impurity : *lowest for best splits*
        - Entropy : *lowest for best splits*
        - Information Gain : *highest for best splits*

4. Split the Data
    - Data is divided into subsets based on the chosen feature.
    - Each subset forms a new branch.
    - Each branch represents a decision outcome.

5. Repeat the Process on Each Branch
    - Steps 2–4 are repeated for every new subset.
    - Each branch keeps splitting until:
        - Data becomes mostly pure, or
        - A stopping condition is reached.

6. Stop the Tree Growth
    - Tree growth stops when:
        - Maximum depth is reached
        - Minimum samples per split condition is met
        - No meaningful impurity reduction is possible

##### What “Best Split” Means
- A good split creates groups that are:
    - More similar within themselves
    - More different from each other
- Measures used to judge split quality:
    - Gini Impurity : *lowest for best splits*
    - Entropy : *lowest for best splits*
    - Information Gain : *highest for best splits*





## 2. Random Forest

`Single Decision Trees are easy to understand but fragile. In real business automations, decisions must be consistent and robust, not sensitive to small data changes. Random Forest solves this by training many trees independently and combining their decisions into one strong outcome.`

##### Core Idea
- Random Forest is an ensemble - supervised learning algorithm.
- It builds multiple Decision Trees instead of one.
- Each tree is trained slightly differently.
- Final prediction is made by combining all trees’ outputs.

##### How Random Forest Works 

1. Create Multiple Random Samples
    - From the original dataset, multiple new datasets are created.
    - Each dataset is formed by random sampling with replacement.
    - This process is called bootstrapping.

2. Train a Decision Tree on Each Sample
    - A separate Decision Tree is trained on each sampled dataset.
    - Sampling happens once per tree, not per iteration.
    - Trees grow independently of each other.

3. Random Feature Selection at Each Split
    - Each tree has access to all features.
    - At every split for each tree, only a random subset of features is considered.
    - Trees are trained independently, not iteratively
    - This prevents trees from becoming too similar.

4. Grow Trees Fully (with Control)
    - Trees are usually grown deep.
    - Overfitting is controlled through randomness rather than pruning.

5. Combine Predictions
    - For classification:
        - Each tree votes for a class.
        - The majority vote is selected.
    - For regression:
        - Predictions from all trees are averaged.

 



## 2. Linear Regression

`In AI automation, many decisions depend on forecasting numbers: revenue, demand, cost, time, or risk. Linear Regression provides a simple and interpretable way to model relationships between inputs and a numeric outcome, making it ideal for early-stage automation and decision support systems.`

##### Core Idea
- Linear Regression is a supervised learning algorithm.
- It models the relationship between input features and a continuous output.

##### How Linear Regression Works 

1. Define Inputs and Output
    - Input features represent influencing factors.
    - Output is a numeric value to be predicted.
    - The goal is to find how changes in inputs affect the output.

2. Assume a Linear Relationship
    - The model starts with an equation combining inputs and weights.
    - Each feature gets a coefficient (weight).
    - A bias term shifts the line up or down.

3. Make Initial Predictions
    - Using initial weights, the model predicts output values.
    - These predictions are usually inaccurate at the beginning.

4. Measure the Error
    - The difference between predicted and actual values is calculated.
    - Mean Squared Error is commonly used to measure this error.

5. Adjust the Line (Learning)
    - The model updates weights to reduce error.
    - This is done using an optimization process.
    - Gradient Descent is commonly used.

6. Repeat Until Best Fit Is Found
    - Steps 3–5 repeat multiple times.
    - Training stops when error cannot be reduced significantly.

7. Make Predictions on New Data
    - Once trained, the model uses the learned line.
    - New input values are plugged in to get predictions.

## 3. Support Vector Machine (SVM) Algorithm 

`In many automation problems, the key challenge is not prediction but clean separation: approve vs reject, fraud vs non-fraud, spam vs non-spam. SVM focuses on finding the most reliable boundary, not just any boundary..`

##### Core Idea

- Support Vector Machine is a supervised learning algorithm.
- It finds a decision boundary that separates data with maximum margin.
- Margin is the distance between the boundary and the closest data points.

##### How Support Vector Machine (SVM) Algorithm Works 

1. Represent Data in Feature Space.
    - Each data point is plotted using its features.
    - The goal is to separate classes using a line or boundary or plane.

2. Find a Separating Boundary
    - Many boundaries may separate the data.
    - SVM evaluates all possible separating boundaries.

3. Maximize the Margin
    - SVM selects the boundary with the largest margin.
    - Margin is measured using the closest data points from each class.

4. Identify Support Vectors
    - The closest data points to the boundary are called support vectors.
    - These points completely define the boundary.

5. Handle Imperfect Separation (Soft Margin)
    - Some data points may lie inside the margin or on the wrong side.
    - A penalty is applied for such violations.
    - A regularization parameter controls strictness.

6. Make Predictions
    - New data is placed in the same feature space.
    - Its position relative to the boundary determines the class.

## Kernel Trick w Support Vector Machines Algorithm

`Real-world business data is rarely linearly separable. Kernel Trick allows SVM to create complex decision boundaries while still using a simple linear separator — making models powerful yet mathematically controlled.`

##### Core Idea

- Some data cannot be separated using a straight line.
- Instead of changing the algorithm, SVM changes how data is viewed.
- Kernel Trick maps data into a higher-dimensional space where separation becomes possible.

##### Types of Kernel Trick

1. Linear Kernel (default)
    - No feature space transformation.
    - Data is linearly separable.

2. Polynomial Kernel 
    - Maps data into polynomial feature space.
    - Can model curved boundaries.

3. Radial Base Function/ Gaussian Kernel
    - Measure similarity based on distance between points.
    - Creates localized influence regions.

    *Gaussian Kernel is a part of RBF Family; however, Gaussian Kernel is highly used in SVM; hence, the terms are inter-changeable. All Gaussial models are RBF but not all RBF are Gaussian.*

4. Sigmoid Kernel
    - Creates S-shaped boundaries.


    *Kernel choice is critical for performance.*


# Difference between Base Functions vs Kernel Trick

`Understanding this helps in deciding whether to explicitly transform features (Base Functions) or let the model handle it implicitly (Kernel Trick) when automating business predictions.`

Similarity : Both Base Functions and Kernel Trick are ways to handle non-linear patterns in data.

Difference : 

| Aspect                              | Base Functions                                                               | Kernel Trick                                                                                         |
| ----------------------------------- | ---------------------------------------------------------------------------- | ---------------------------------------------------------------------------------------------------- |
| **Definition**                      | Explicitly transform input features  | Implicitly compute similarity in a higher-dimensional space without transforming features explicitly |
| **Transformation**                  | Done manually or programmatically                                            | Done mathematically inside the algorithm using kernels                                               |
| **Feature Space**                   | You generate new features (e.g., polynomial, Gaussian, sigmoid)              | No explicit features are generated; only inner products (similarity) are computed                    |
| **Memory / Computation**            | Can be expensive if feature space is high-dimensional                        | Often more efficient, avoids storing high-dimensional data explicitly                                |
| **Model Applicability**             | Linear models can handle non-linear patterns after feature transformation    | Primarily used in kernelized algorithms like SVM, K Means.                                                     |
| **Transparency / Interpretability** | Easy to inspect transformed features                                         | Hard to inspect; transformation is implicit                                                          |
| **Examples**                        | Polynomial basis, Gaussian basis, Sigmoid basis                              | Gaussian RBF kernel, Polynomial kernel, Sigmoid kernel in SVM                                        |


## 4. K Means Clustering Algorithm

`K-Means helps in automatically segmenting customers, leads, transactions, or behaviors without labels — critical for automation and decision-making.`

##### Core Idea

- K-Means is an unsupervised learning algorithm.
- It groups similar data points into K clusters (groups).
- Group things so similarity within the group is maximum and similarity inbetween the clusters is minimum.
- It is an iterative process.
- K-Means does not learn rules, it learns structure.

##### How K Means Clustering Algorithm Works 

1. Decide the number of clusters (K)
    - K represents how many groups you want in the data.
    - In order to find optimal values of K, Silhouette Score, Elbow Method and other analysis methods has been used.

2. Initialize K centroids (Expectation)
    - Randomly place K points in the feature space. 
    - Each centroid represents the center of a cluster.

3. Measure distance from each data point to each centroid. 
    - Distance is usually calculated using Euclidean distance.
    - This measures similarity between data points and centroids.
    - Similarity is inversely proportional to distance.

4. Assign each data point to the nearest centroid *(Expectation)*
    - Each point becomes part of the cluster with minimum distance
    - This forms temporary clusters.

5. Recalculate centroid positions *(Maximization)*
    - For each cluster, compute the mean of all assigned points.
    - Move centroid to this new mean position.

6. Repeat assignment and update steps
    - Reassign points based on updated centroids.
    - Recompute centroids again.

7. Check for convergence
    - Stop when centroids stop changing
    - Or when changes are very small
    - Or when maximum iterations are reached

## 5. Gaussian Mixture Models 

`As an AI Automation Business Consultant, GMM lets you model real-world data where groups overlap and decisions must be probabilistic, not rigid.`

##### Core Idea

- Gaussian Mixture Model (GMM) is a probabilistic unsupervised clustering algorithm.
- Data is modeled as a combination of multiple Gaussian distributions.
- Uses Expectation–Maximization (EM) algorithm just like K-Means:
    - E-step: estimate probability of each point belonging to each cluster
    - M-step: update Gaussian parameters 

##### How Gaussian Mixture Models Clustering Algorithm Works 

1. Choose number of components (K) (*here we call components and not clusters*).
    - Decide how many Gaussian distributions will model the data.
    - Value of K would be calculated with the help of AIC, BIC or more.
2. Random initialisation of Gaussians.
3. Mapping of each data point to the Gaussian distribution based on the probability of beloningness.  (*Expectation*)
4. Recalculate Gaussian distributions. (*Maximisation*)
5. Repeatition of step 3 and 4 till convergence.

## 6. Naive Bayes Models 

`Naive Bayes is used for fast, scalable classification, especially in text-heavy business automation like spam detection, ticket routing, and intent classification.`

##### Core Idea

- Naive Bayes is a supervised probabilistic classification algorithm.
- It predicts how often a feature would beloong to a particular class.
- Extremely fast and works well with high-dimensional data.
- It assumes that all features are conditionally independent.
- Types of Naive Bayes Models : 
    - Gaussian Naive Bayes → continuous numerical data
    - Multinomial Naive Bayes → text data, word counts
    - Bernoulli Naive Bayes → binary features (yes/no)

##### How Naive Bayes Algorithm Works 

**Easy to Understand**

- Feature Likelihood: How often a feature appears for a given class (e.g., red apples = 70%)
- Class Prior: How common each class is in the dataset (e.g., apples = 50%)
- Posterior for unseen sample: Combine prior and likelihoods to predict class (e.g., new red small round fruit → most likely apple)
- Prediction: Class with highest posterior probability is chosen

**Steps**

1. Collect and prepare training data.
2. Calculate class priors (P(Class))
    - Count how frequent each class occurs in the training data.
3. Calculate feature likelihoods.
    - For each class, compute probability of each feature value.
4. Train the model.
5. Store class priors and feature likelihoods for later prediction.
6. Predict for unseen data using class priors and feature likelihoods. 
7. Choose the class with highest posterior probability.
    - Assign this class as the predicted label for the new sample.

# Ensemble Learning

`As an AI Automation Business Consultant, ensemble learning helps you deliver more accurate and stable AI systems for clients where failure is costly by combining outputs of multiple machine learning models rather than just relying on one.`

##### Methods :

1. Voting.
    - Works mainly for classification.
    - Two main types:
        - Hard Voting: final class is the majority vote.
        - Soft Voting: average predicted probabilities, then choose highest.
2. Averaging.
    - Mainly used for regression problems.
    - Simple average (mean) of output by ML models would be the final output.
3. Weighted Average.
    - Mainly used for regression problems.
    - Explicit biasedness has been given specific models.
    - Assigns more trust to strong models, less to weaker ones. 
4. Stacking.
    - Used for both classification and regression problems. 
    - Multiple ML models could be collaborated in order to determine final output. 
    - Base model (Level 0) is being trained on the original dataset.
    - Base model (Level 0) predictions would be used to train meta model (Level 1).
    - Meta-model learns how to best combine predictions to produce final output.
    - Cross validation testing is done at level 0.
5. Blending.
    - Similar to Stacking; however, provides more customisation.
    - Meta models are called blender models.
    - hold out validation testing is done at level 0.
6. Bagging. 
    - Bagging (Bootstrap Aggregating) is an ensemble learning technique.
    - Multiple models are trained independently and in parallel.
    - Each model is trained on a bootstrap sample (random sampling of features and dataset with replacement).
    - Final prediction is obtained using:
        - Voting for classification.
        - Averaging for regression.
7. Boosting.
    - Used for classification problems. 
    - It is an iterative algorithm. 
    - Boosting is an ensemble technique where models are trained sequentially.
    - Each new model focuses more on previously misclassified or poorly predicted data.
    - Final prediction is a weighted combination of all models.
    - Common boosting algorithms : 
        - AdaBoost.
        - Gradient Boosting.
        - XGBoost, LightGBM, CatBoost.
