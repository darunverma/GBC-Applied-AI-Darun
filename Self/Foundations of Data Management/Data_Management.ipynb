### Normalisation

It is a database designing technique of decomposing large tables into small tables in order to reduce (minimise) redundancy and inconsistencies. 

**Normalisation Forms :**

1. 1NF
    - Each record/ instance is uniquely identified (*combination of key values must be unique*)
    - No repeating groups/ arrays/ attributes.
    - Each attribute must contain atomic values.
2. 2NF
    - 1NF
    - No partial dependencies (*no non-key attribute is dependent on a part of primary key and not whole key.*)
3. 3NF
    - 2NF
    - No transitive dependencies (*no non-key attribute is dependent on another non-key attribute which is further depended on primary key.*)
4. BCNF
    - 3NF
    - No non-prime attribute derives the value of a prime attribute.

*Normalisation helps to avoid : Delition Anomly, Updation Anomly, Insertion Anomly.*

## 6. Data Storage & Operations

Data must be stored correctly, accessed efficiently, and maintained consistently — this is the operational promise of this knowledge area.

**Principles of Data Storage and Operations :**

1. Data Availability Principle :    
    - Data must be accessible to authorized users and systems at all times. Downtime is a business cost.
2. Data Integrity Principle :    
    - Data stored must be consistent, accurate, and trustworthy. Corrupted data produces corrupted AI outputs.
3. Manageability Principle :    
    - Storage systems must be manageable — monitored, maintained, and updated without disrupting operations.
4. Scalability Principle :    
    - Storage architecture must grow with the business — handling increasing data volumes without performance degradation.
5. Performance Principle :    
    - Read and write operations must be optimized — through indexing, caching, and efficient query design.
6. Security and Access Control Principle :    
    - Only authorized roles should access specific data — enforced through permissions, encryption, and audit trails.
7. Retention and Disposal Principle :    
    - Data must be retained for the required period and securely disposed of when no longer needed — legally and operationally.
8. Recovery and Continuity Principle :    
    - Disaster recovery plans and backup strategies must be in place to ensure business continuity.

**Database Architecture Types :**


| Feature | Centralized | Distributed Non-Federated | Distributed Federated |
|---|---|---|---|
| **Data Location** | Single location | Multiple nodes (sharded) | Multiple autonomous systems |
| **DBMS** | Single DBMS | Single DBMS managing all nodes | Multiple diverse DBMS (SQL, NoSQL, Legacy) |

| **So Instead** | One big safe for all the company's gold | A chain of identical stores managed by one central headquarters | A translator who speaks to 10 different people and gives you one combined answer |

*Virtualization is the software that splits one physical server into multiple digital ones, while Cloud Platforms are the internet-based services that let you rent those digital servers on demand.*


**Database Processing Types :**

**ACID (Structred Data)**

- Atomicity — Every transaction fully completes or fully rolls back — never half-done
- Consistency — Data always moves from one valid state to another after every transaction
- Isolation — Concurrent transactions do not interfere with each other
- Durability — Committed data is permanently saved even after system failure

So instead: A bank transfer that either fully goes through or fully cancels — never stuck in the middle.

**BASE (UnStructred Data)**

- Basically Available — System always responds even if data is slightly outdated or incomplete
- Soft State — System state can change over time even without new input — not always stable
- Eventual Consistency — Data will become consistent and accurate across all nodes — just not immediately

So instead: A social media like counter that shows 1.2k instead of exactly 1,247 — close enough, always available.

**CAP Theorem (Distributed)**

- Consistency — All nodes in the system see the same data at the same time
- Availability — Every request receives a response, even if it is not the most current data
- Partition Tolerance — System continues operating even if network communication between nodes fails

The Rule — A distributed system can only fully guarantee 2 of these 3 properties simultaneously.

So instead: You are making 3 promises but reality only lets you keep 2 — you must consciously choose which one to compromise.

## 7. Data Quality

Data Management is actually managing the quality of data throughout its lifecycle to make sure it is suitable for its intended usage.

*Quality of data is highly subjective and based on context.*

**Data Quality Improvement Lifecycle :**

*PLAN → DO → CHECK → ACT → Repeat*

1. PLAN: Identify known data issues and define standards.
2. DO: Remediate (fix) the issues.
3. CHECK: Monitor the data to ensure quality is maintained.
4. ACT: Address new or emerging data issues discovered during monitoring.

**Data Profiling (Inspection):**

Analysis of data to assess structure and quality.

**Data Processing (Improvement):**

Improving the quality of data to make it usable.

It includes:
- Cleaning.
- Transforming.
- Aggregating.
- Enriching.
- Validating.

**Preventive vs Corrective Actions :**

Preventive : Stop "garbage" from entering the system.      
Corrective : Fix "garbage" that is already in the system.

**Measuring Data Quality :** 

| Metric / Dimension             | Formula                                                | Unacceptable Threshold              |
| ------------------------------ | ------------------------------------------------------ | ----------------------------------- |
| **Validity Ratio**             | (Test Executions − Exceptions Found) ÷ Test Executions | Lower value indicates poor quality  |
| **Invalidity Ratio**           | Exceptions Found ÷ Test Executions                     | Higher value indicates poor quality |
| **Completeness**               | (Populated Records ÷ Total Records) × 100              | < 80%                               |
| **Uniqueness**                 | (Duplicate Records ÷ Total Records) × 100              | > 0%                                |
| **Timeliness**                 | (Delayed Transactions ÷ Total Transactions) × 100      | On-time rate < 99%                  |
| **Validity (Rule Compliance)** | (Records Meeting Rules ÷ Total Records) × 100          | < 99%                               |


*It is helpful to build Service Level Agreements (SLA).*

**Difference :**

| Feature           | Operational Database (OLTP)          | Data Warehouse (OLAP)                      |
| ----------------- | ------------------------------------ | ------------------------------------------ |
| Primary Purpose   | Run daily transactions               | Analytical reporting & BI                  |
| Processing Type   | OLTP (Online Transaction Processing) | OLAP (Online Analytical Processing)        |
| Users             | Clerks, applications, APIs           | Analysts, executives, data scientists      |
| Data Type         | Current operational data             | Historical, integrated data                |
| Data Updates      | Frequent insert/update/delete        | Mostly append-only                         |
| Volatility        | Highly volatile                      | Non-volatile (stable)                      |
| Data Structure    | Highly normalized                    | Often denormalized (Star/Snowflake schema) |
| Query Type        | Simple, short transactions           | Complex aggregations & joins               |
| Query Performance | Optimized for fast writes            | Optimized for fast reads                   |
| Data Volume       | Moderate (current records)           | Very large (years of history)              |
| Time Focus        | Present state                        | Time-variant (historical trends)           |
| Source of Data    | Single application                   | Multiple integrated sources                |
| Concurrency       | Thousands of simultaneous users      | Fewer users, heavy queries                 |
| Examples          | Banking system, POS, CRM             | Enterprise reporting system                |
